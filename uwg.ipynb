```python
from uwg import UWG

# Change this to the full path of your Singapore EPW file
epw_path = r"E:\ALi_Vatan\Python\urbanWeatherGen-master\resources\epw\SGP_Singapore.486980_IWEC.epw"   # or full path like r"C:\Users\YourName\Downloads\SGP_Singapore.486980_IWEC.epw"

# Create the model (this uses Singapore defaults, same as your initialize_singapore.uwg)
model = UWG.from_param_args(
    epw_path=epw_path,
    bldheight=10,      # average building height in meters
    blddensity=0.5,    # building density
    vertohor=0.8,      # vertical to horizontal ratio
    grasscover=0.2,
    treecover=0.1,
    zone='1A',         # hot-humid climate (Singapore)
    new_epw_name="SGP_Singapore_Urban_Test.epw",

    month=1,      # Starting month (1 = January)
    day=1,        # Starting day
    nday=32,      # ←←← Change this number (e.g. 7 for one week, 15, 31, 60, etc.)

)

print("Running Urban Weather Generator for Singapore...")
model.generate()
model.simulate()
model.write_epw()

print("✅ Finished! New urban EPW file created: SGP_Singapore_Urban.epw")
```
